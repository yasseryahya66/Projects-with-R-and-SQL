Write a function is_even(number)

Returns "Even" if the number is even

Returns "Odd" otherwise

```python
def is_even(number):
    if (number // 2) * 2 == number:
       return "even"
    else:
       return "odd"


# apply the function
is_even(5)
```

Write a function number_sign(n)

"Negative" if n < 0

"Zero" if n == 0

```python
def number_sign(n):
  if n > 0:
    print("Positive")
  elif n < 0:
    print("Negative")
  elif n == 0:
    print("Zero")

# Apply the funtion
number_sign(-1)
number_sign(1)
number_sign(0)
```

Write a function get_grade(score)

90–100 → "A"

80–89 → "B"

70–79 → "C"

60–69 → "D"

Below 60 → "F"

```python
def get_grade(score):

  if score >= 90:
         return "A"
  elif score >= 80:
         return "B"
  elif score >= 70:
         return "C"
  elif score >= 60:
         return "D"
  else:
         return ("F")


# Apply function
print(get_grade(50))
print(get_grade(60))
print(get_grade(70))
print(get_grade(80))
print(get_grade(90.5))

```

Write a function max_of_two(a, b)

Returns the larger number

If they are equal, return "Equal"

```python
def max_of_two(a,b):
  if a == b:
        return "Equal"
  else:
        return max(a,b)


# Apply the function
print(max_of_two(1,1))
print(max_of_two(10,5.5))
```

Write a function weather_advice(temp)

Below 0 → "Freezing"

0–15 → "Cold"

16–25 → "Mild"

Above 25 → "Hot"


```python
def weather_advice(temp):
  if temp > 25:
          return "Hot"
  elif temp >= 16:
          return "Mild"
  elif temp >= 0:
          return "Cold"
  else:
          return "Freezing"

# apply the function
print(weather_advice(26))
print(weather_advice(16))
print(weather_advice(2))
print(weather_advice(-1))
```

"Login successful" if both match

Write a function is_leap_year(year)
A year is a leap year if:

Divisible by 4

Not divisible by 100 unless also divisible by 400
Return True or False.

```python
def is_leap_year(year):
  if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
  else:
        return False


# Apply the function
print(is_leap_year(2020))  # True
print(is_leap_year(1900))  # False
print(is_leap_year(2000))  # True
print(is_leap_year(2021))  # False
```

Write a function calculator(a, b, operator)
Operators:

"+" → addition

"-" → subtraction

"*" → multiplication

"/" → division (handle division by zero!)

Return "Invalid operator" if operator isn’t supported.

```python
def calculator(a, b, operator):
    if operator == "+":
        return a + b
    elif operator == "-":
        return a - b
    elif operator == "*":
        return a * b
    elif operator == "/":
        return "Cannot divide by zero" if b == 0 else a / b
    else:
        return "Invalid operator"


# Apply the function
print(calculator(5,0, "/"))
print(calculator(5,0, "+"))
print(calculator(5,0, "*"))
print(calculator(5,0, "-"))
```
